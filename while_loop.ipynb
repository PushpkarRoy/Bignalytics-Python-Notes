__Q1. ATM Withdrawal System (While Loop)__

    A user keeps withdrawing money until:

__Q2. Login System__

    A system allows only 3 attempts to enter the correct password.

__Rules__

    Correct password = "admin123"
    Lock account after 3 failed attempts
    Print success or lock message

```python
password = "admin123"

trying = 0 

while trying < 3:
    user_password = input("Enter your password:- ")
    if password == user_password:
        print("Thank you for login")
        break
    elif password != user_password:
        print("Your password is incorrect please try again")
        trying +=1
    if trying == 3:
        print("Your account is lock please try again later") 
```

__Q3. Daily Sales Tracking__

    Keep taking daily sales until the user enters -1.

__Output__

    Total sales
    Number of days
    Average sale

```python
sales = int(input("Enter your sales :- "))
total_sales = 0
days = 1

while sales != -1:
    total_sales = sales + total_sales
    sales = int(input("Enter your sales :- "))
    days += 1
print("Your total sales amount is :-", total_sales)
print("Total days is :- " ,days - 1)
avg_sales = total_sales // (days -1)
print("Avg sales is :- ", avg_sales)
```

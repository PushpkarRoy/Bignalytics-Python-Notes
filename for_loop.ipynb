__Q4. A shop records daily sales for 7 days:__

Write a program to:

    calculate total sales

    calculate average sales

sales = [1200, 1500, 900, 1800, 2000, 1100, 1600]

```python
sales = [1200, 1500, 900, 1800, 2000, 1100, 1600]

total_sales = 0
total_days = len(sales)

for i in sales:
    total_sales = i + total_sales
    avg_sales = total_sales / total_days
print("Total_sales is :- ", total_sales)
print("Avg sales is :- ",  avg_sales)

```

__Q5. You have a list of student marks:__

marks = [45, 67, 89, 32, 76, 50, 90]

Print:

        "Pass" if marks ≥ 50

        "Fail" if marks < 50

```python
marks = [45,67,89,32,76,50,90]
pass_students = 0
fail_students = 0

for i in marks:
    if i >= 50:
        print(i,"Pass")
        pass_students += 1
    else:
        print(i,"fail")
        fail_students += 1

print("Total pass students is :- ", pass_students)
print("Total fail students is :- ", fail_students)
```

Question 10. Write a program to count occurrences of each character in the word. "Programming".

```python
letter = {}
word = "programming"


for i in word:
    if i in letter:
        letter[i] += 1
    elif i not in letter:
        letter[i] = 1


print(letter)

    
```

<h2 style="color:darkorange; font-weight:bold;">Nested if___</h2>

__🔹 What is Nested if?__

Nested if = if ke andar ek aur if

        Matlab condition ke andar condition

        Execute hota hai outer if true hone par hi inner check hoti hai

__Example__

```python

# if condition1:
    # if condition2:
        # action
```

__Why Nested if?__

        Jab decision multi-level ho

        Multiple conditions sequential check karni ho

__Real-world example:__

        Bank loan eligibility

        College admission

        Login + 2FA check

```python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to vote")
else:
    print("Not Eligible for vote")
```

__Flow:__

Outer if → check age

Inner if → check citizenship

Both True → print statement

```python
# Real-World Example 1: Bank Loan

salary = 50000
credit_score = 750

if salary >= 40000:
    if credit_score >= 700:
        print("Loan Approved")
    else:
        print("Improve credit score")
else:
    print("Salary too low")
```

```python
# Real-World Example 2: Exam Result

marks = 85
attendance = 75

if marks >= 40:
    if attendance >= 75:
        print("Pass")
    else:
        print("Fail due to low attendance")
else:
    print("Fail due to low marks")
```

__Rules to Remember__

        Indentation matters

        Outer if must be True for inner if to check

        Can use else and elif inside nested structure

__🔹 Common Mistakes__

        Forgetting inner indentation → IndentationError

        Confusing inner and outer logic → wrong output

        Too many nested levels → code hard to read

```python
# Nested if
a = 1
b= 5

if a > 0:
    if b > 0:
        print("Both positive")

# Multiple if
if a > 0 and b > 0:
    print("Both positive")

# ✅ Nested if → readable for multi-step decisions
# ✅ Multiple if → simpler for combined condition
```

__Use of * and ** in Functions__

    1️⃣ *args (Single Star)
    🔹 What is *args?

        *args allows a function to:
        accept multiple positional arguments
👉 Jab arguments ki count fixed na ho

```python
def add(*args):
    total = 0
    for num in args:
        total += num
    return total

add(1, 2, 3, 4)

# 📌 Here:
# args = (1, 2, 3, 4)
# Type → tuple
```

```python
# 🔹 Why use *args?

# Without it ❌:

def add(a, b, c, d):
    ...


# With it ✔️:

def add(*args):
    ...
```

__Rules of *args__

    args is just a name (standard)
    Always comes after normal parameters

```python
def test(a, b, *args):
    ...
```

__What is OOPs?__

Definition:

    Object-Oriented Programming (OOP) is a programming paradigm that organizes code into objects which have attributes (data) and methods (functions) to model real-world entities.

Simple words:

    Real-world ko code me model karna
    Objects = things / entities
    Classes = blueprint for objects

Analogy:

    Class = blueprint of a car
    Object = actual car

__Key Features of OOPs in Python__

__Feature__	-------------------__Definition__ ------------------------------------------__Example__

Class	----------------------Blueprint for objects	--------------------------------------class Car:

Object------------------------Instance of class------------------------------------------my_car = Car()

Encapsulation-----------------Hiding internal details ------------------------------------Private variables

Inheritance -----------------Reuse properties/methods from another class ------------------class ElectricCar(Car):

Polymorphism------------------Same function, different behavior ----------------------------len() works on list & string

Abstraction ------------------Hide complexity, show only necessary------------------------Abstract class / method

```python
# 🔹 3️⃣ Classes & Objects
# Class Syntax


class Car:
    wheels = 4           # Class attribute
    
    def __init__(self, brand, color):   # Constructor
        self.brand = brand              # Instance attribute
        self.color = color
    
    def drive(self):                     # Method
        print(f"{self.brand} is driving")
```

```python
# Create Object
my_car = Car("BMW", "Black")
print(my_car.brand)     # BMW
my_car.drive()          # BMW is driving


# Notes:
# self = reference to current object
# __init__ = constructor, automatically called during object creation
```

```python
# Attributes

# 1. Class Attribute → Shared among all objects

class Car:
    wheels = 4


# 2. Instance Attribute → Unique per object

my_car.color = "Black"


# 3. Private Attribute → Encapsulation

class Car:
    def __init__(self):
        self.__engine = "V6"   # private
```

__Methods__

```python
# 1. Instance Method → Works with object

def drive(self):
    print("Car is driving")
```

```python
# 2. Class Method → Works with class, not instance

@classmethod
def car_info(cls):
    print("All cars have", cls.wheels, "wheels")
```

```python
# 3. Static Method → Independent of class & object

@staticmethod
def welcome():
    print("Welcome to Car world")
```

__Inheritance__

```python
# Definition:

# Child class inherits properties and methods of parent class

class ElectricCar(Car):   # Inheritance
    def charge(self):
        print("Charging battery")
```

## Polymorphism

__Definition:__

Same function / operator behaves differently in different scenarios

```python
# Example 1: Method Overriding

class Car:
    def drive(self):
        print("Car driving")
        
class ElectricCar(Car):
    def drive(self):
        print("Electric car driving silently")
```

```python
# Example 2: Operator Overloading

print(2 + 3)       # addition
print("Hi" + "Bye") # concatenation
```

## Abstraction

__Definition:__

Hide unnecessary details and show only important info

```python
# Using abstract class

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def move(self):
        pass


# Child class must implement move()
```

__Special Methods / Magic Methods__


 __Method__	----------------------------------------------__Use__

    __init_--------------------------------------Constructor
    __str__----------------------------------------Human-readable object print
    __len__-----------------------------------------Return length
    __add__----------------------------------------Overload + operator

```python
class Book:
    def __init__(self, pages):
        self.pages = pages
        
    def __str__(self):
        return f"Book with {self.pages} pages"

b = Book(200)
print(b)   # Book with 200 pages
```

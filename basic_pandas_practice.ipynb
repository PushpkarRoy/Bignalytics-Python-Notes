## 🟡 Level 2 – GroupBy Practice (Very Important 🔥)
__6️⃣ Department Analysis__

Using employee DataFrame:

    Find average salary per department
    Find highest salary per department
    Count employees per department

```python
# Find average Salary per department 
employes.groupby("Department")["Salary"].mean()
```

```python
# Find highest Salary per department 
employes.groupby("Department")["Salary"].max()
```

```python
# Find highest Salary per department    # Other eay 
employes.loc[employes.groupby("Department")["Salary"].idxmax()]
```

```python
# Count employees as per the department 
employes.groupby("Department")["Department"].count()
```

7️⃣ Real-World: Sales Dataset

Create DataFrame:

| OrderID | Product | Quantity | Price |

Then:

    Create new column: Total = Quantity * Price
    Find total revenue
    Find product with highest total sales
    Find average order value

```python
data = {
    "OrderID":[101,102,103,104,105,106,107,108,109],
    "Product":["Apple","Banana","Mango","Apple","Apple","Banana","Orange","Mango","Mango"],
    "Quantity":[10,12,20,5,12,12,7,12,15],
    "Price":[20,5,25,20,20,5,15,25,25]
}
```

```python
shop = pd.DataFrame(data)
```

```python
shop
```

```python
# Create new column: Total Revenue = quantity * price 
shop["Total Revenue"] = shop["Quantity"] * shop["Price"]
shop
```

```python
# Find product with the highest total sales 
shop.groupby("Product")["Total Revenue"].sum()
```

```python
# Find avg order value 
total_revenue = shop["Total Revenue"].sum()           # Total Revenue / Total order = AOV (Avg order value)
total_order = shop["OrderID"].count()

aov = total_revenue / total_order
aov 
```

```python
# Find avg order value               # Single step method 
shop["Total Revenue"].mean()
```

## 🟠 Level 3 – Slightly Logical
__9️⃣ Remove Duplicates__

Create DataFrame with duplicate rows.

    Remove duplicates
    Reset index

```python
import pandas as pd
import numpy as np  

data = {
    'Employee_ID': [101, 102, 103, 101, 104, 105, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Eve', 'Eve'],
    'Department': ['HR', 'IT', 'Marketing', 'HR', np.nan, 'Sales', 'Sales'],
    'Salary': [50000, 60000, np.nan, 50000, 55000, 70000, 70000]
}
```

```python
comp = pd.DataFrame(data)
comp
```

```python
# Find and remove duplicates 

comp.duplicated()
```

```python
comp[comp.duplicated()]
```

```python
# Remove Duplicates 
comp = comp.drop_duplicates()
```

```python
comp
```

```python
# Reset Index 
comp.reset_index()
comp
```

```python
comp.isnull().sum()
```

```python
comp.fillna("I don't know")
```

```python
# How to fill salary missing value 
comp["Salary"].fillna(comp["Salary"].mean(), inplace= True)
```

```python
comp 
```

```python
# Fill Department missing value 
comp["Department"].fillna("IT", inplace= True)
```

```python
comp
```

🔟 Apply Function

Create a function:

    If Salary > 70000 → "High"
    If 40000–70000 → "Medium"
    Else → "Low"

```python
salary = comp["Salary"]
salary
```

```python
salary_slab = []
for i in salary:
    if i >= 70000:
        salary_slab.append("High")
    elif i >= 52000 and i < 70000:
        salary_slab.append("Medium")
    else:
        salary_slab.append("Low")
print(salary_slab)
```

```python
comp["Salary_Level"] = salary_slab
```

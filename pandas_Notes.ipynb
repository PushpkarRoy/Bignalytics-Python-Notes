## What is Pandas 

Pandas is an open-source Python library designed specifically for data manipulation and structured data analysis. It was created by Wes McKinney to solve real-world financial data problems, and today it has become the backbone of data analysis in Python.

At its core, Pandas provides high-level data structures and tools that allow us to work with structured datasets efficiently. It is built on top of NumPy, which means it inherits NumPy’s speed and array-processing power but adds labeled axes (rows and columns), making data much easier to understand and manipulate.

Unlike plain Python lists or dictionaries, Pandas is optimized for:

    Tabular data (rows and columns)
    Handling missing values
    Performing statistical operations
    Working with time series data
    Reading and writing different file formats

The two primary data structures in Pandas are:

    Series (1D labeled array)
    DataFrame (2D labeled data structure)

In simple terms:

    Pandas = Powerful tool to clean, transform, and analyze structured data efficiently.

<h2 style="color:darkorange; font-weight:bold;">Data Frame</h2>

A DataFrame is the most important object in Pandas. It is a two-dimensional, size-mutable, heterogeneous tabular data structure with labeled axes.

Let’s break that definition carefully.

    Two-dimensional:
    It has rows and columns.

    Size-mutable:
    You can add or remove rows and columns dynamically.

    Heterogeneous:
    Each column can have different data types.

    Labeled axes:
    Both rows and columns have labels.

Internally, a DataFrame is built from multiple Series objects. Each column in a DataFrame is a Series. So technically:

    DataFrame = Collection of aligned Series sharing the same index.

This alignment feature is very powerful. If you perform operations between two columns, Pandas aligns them based on index automatically.

Example structure conceptually:

    Index → Row labels
    Columns → Column labels
    Values → Data stored in NumPy arrays internally

Internally, Pandas stores data in blocks for performance optimization. Numeric columns are often grouped together for vectorized computation.

Key properties of a DataFrame:

    It is built on NumPy arrays.

    It supports vectorized operations.

    It automatically handles missing values.

    It supports powerful indexing mechanisms.

This makes DataFrame far superior to basic Python lists for data analysis.

<h2 style="color:darkorange; font-weight:bold;">Rows & Columns - Selection</h2>

__Understanding Selection in Pandas__

In Pandas, selecting rows and columns is one of the most fundamental and frequently used operations. Every data analysis task involves extracting specific parts of data based on business requirements. Unlike Excel, where we manually select cells, Pandas provides powerful indexing and filtering mechanisms that are programmatic, scalable, and extremely flexible.

[]

.loc[]

.iloc[]

Each has a specific purpose.

```python
# Select Columns 
load_df[["Name"]]
load_df[["Name","Monthly_Salary"]]
```

Rows are selected differently. Pandas does not directly allow row selection using simple bracket notation.

For rows, we use:

    .loc[] → Label-based selection
    .iloc[] → Position-based selection

```python
load_df.loc[0:3]   # [0:3] they include all value from 0 to 3. 3 is also included 
                    # .loc[] selects data based on row and column labels.
```

```python
# Select Rows
# loc - index name based
load_df.loc[load_df.Name == "Pushpkar Roy"]  # Using loc formula to excess the Row and using [].

# using double filter 

load_df.loc[(load_df.Name == "Ankit Roy") & (load_df.Age >= 20)]
```

```python
# iloc = index value baded 

load_df.iloc[0]    # 0 is the index value mean 1 value of the table 
load_df.iloc[0:3]  # [o:3] They show all value between 0 to 3. 3 is not included 
                    #[start:stop:step]
                    # .iloc[] selects data based on numerical position.
```

__Difference Between loc and iloc (Core Concept)__

Feature_____________________________loc___________________________iloc

Based On___________________________Labels_________________________Position

Slice End_________________________Inclusive________________________Exclusive

Works With_______________________Index names_______________________Index numbers

<h2 style="color:darkorange; font-weight:bold;">Concatenate and merge Dataframe (JOINS)</h2>

```python
df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["A","B","C"]
})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    "ID":[1,2,2,4],
    "Score":[88,96,77,79]
})

df2
```

## How to concat 2 or or then 2 tables

```python
pd.concat([df1,df2], axis= 0)  # Vertical joint / row level join / top-on-top join 
```

```python
pd.concat([df1,df2], axis= 1) #Horezontal joint / column joint / side on side joint 
```

## How to merge or join the table 

```python
pd.merge(df1,df2, how= "inner", on= "ID")   # Just like  a SQL first left table then right table 
                                            # Then how we join inner, outer, side, left, right, full 
                                            # then using ON mean in witch condetion or common column in both the table like SQL 
```

```python
# How to use left join
pd.merge(df1,df2, how= "left", on= "ID")
```

```python
# How to use right join 
pd.merge(df1,df2, how="right",on="ID")  # After typing how= python aotomaticaly show all type of join you use according to the work 
```

__What if the On column name is defferent in both the tables__

```python
pd.merge(df1,df2, how="inner", left_on= "ID", right_on= "ID")  # we are using left_on and right_on if the name of the column is different in both the table 
```

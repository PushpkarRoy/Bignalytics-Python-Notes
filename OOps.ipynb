```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def welcome(self):
        print("welcome student", self.name)
    
    def show_marks(self):
        print("Your marks is ",self.marks)

s1 = Student("Monty",77)
print(s1.name,s1.marks)
s1.welcome()
s1.show_marks()
```

```python
s2 = Student("Karan", 25)
print(s2.name,s2.marks)
```

```python
# Student and marks class 


class Students:
    def __init__(self,name,hindi,english,maths):
        self.name = name 
        self.hindi = hindi
        self.english = english
        self.maths = maths 

    def show_marks(self):
        print("Welcome student", self.name)
        total_marks = self.hindi + self.english + self.maths
        avg_marks = (self.hindi + self.english + self.maths) / 3 
        return "Your total marks is ", total_marks, "And your Avg score is ", avg_marks

    
st1 = Students("Pushpkar", 77,88,99)
st1.show_marks()
print(st1.hindi)
st1.english
st1.maths
```

```python
st2 = Students("Pihu", 50,50,50)
st2.show_marks()
```

Create a class BankAccount with:

    Attributes: name, balance
    Methods:
        deposit(amount) → add money
        withdraw(amount) → subtract money (only if balance is enough)
        show_balance() → print current balance

Extra: Use while loop to let user do multiple transactions until they type “exit”.

```python
class Bank_account:
    def __init__(self,name,balance):
        self.name = name 
        self.balance = balance 
    def credit(self, amount):
        self.balance += amount
        print("Thank you for the deposit...")
        print(amount,"Rs was credit in your account")
        print("Your currnt balance is ", self.balance)
    
    def debit(self, amount):
        self.balance -= amount
        print("Thank you for the withdraw")
        print(amount,"Rs amount debit from your account")
        print("Your currnt balance is ", self.balance)
    
    def current_balance(self):
        print("Your final balance is ", self.balance)




us1 = Bank_account("Monty",10000)
        
```

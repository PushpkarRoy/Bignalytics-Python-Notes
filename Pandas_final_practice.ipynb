```python
import pandas as pd 
```

```python
data = {
    "Order_ID": [101,102,103,104,105,106,107,108,109,110],
    "Customer": ["Amit","Riya","Amit","Neha","Rohan","Riya","Neha","Amit","Rohan","Riya"],
    "Product": ["Laptop","Mouse","Keyboard","Laptop","Monitor","Keyboard","Mouse","Laptop","Mouse","Monitor"],
    "Category": ["Electronics","Accessories","Accessories","Electronics","Electronics","Accessories","Accessories","Electronics","Accessories","Electronics"],
    "Price": [60000,500,1500,62000,12000,1400,550,61000,600,12500],
    "Quantity": [1,2,1,1,1,3,2,1,5,1],
    "City": ["Delhi","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Mumbai","Delhi","Pune"],
    "Rating": [4.5,4.0,3.8,4.7,4.2,3.9,4.1,4.8,3.7,4.3]
}

df = pd.DataFrame(data)
df
```

__Q1. Show all orders where price is greater than 10,000.__

```python
df[df["Price"]> 10000]
```

__Q2. Find orders where:__

    Quantity greater than 2

```python
city = df[df["City"] == "Delhi"]
city[city["Quantity"] > 2]
```

__Q3. Create a new column called:__

    Total_Amount

```python
df["Total_Amount"] = df["Price"] * df["Quantity"]
df
```

__Q4.  Sort the dataframe by:__

    Total_Amount (descending)
    Show top 5 rows.

```python
df.sort_values(by="Total_Amount", ascending= False).head(5)
```

__Q6. Find:__

    Total Sales by Product

```python
df.groupby("Product")["Total_Amount"].sum()
```

__Q7. Find:__

    Average Rating for each Category
    Sort result highest rating first

```python
round(df.groupby("Product")["Rating"].mean().sort_values(ascending= False),2)
```

__Q8. Find the customer who spent the most money.__

    Output should show:
    Customer | Total Spending

```python
df.groupby("Customer")["Total_Amount"].sum().sort_values(ascending=False).reset_index()
```

__Q9. Find which city generated the highest revenue.__

```python
df.groupby("City")["Total_Amount"].sum().sort_values(ascending= False).reset_index()
```

__Q10. Find:__

    Top 3 most sold products
    Based on total quantity sold.

```python
df.groupby("Product")["Quantity"].sum().sort_values(ascending= False).reset_index().head(3)
```

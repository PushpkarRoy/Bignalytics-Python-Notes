__Q10. Function Practice (5 Marks)__

    Write a function is_prime(num)
    Returns True if prime, False otherwise

```python
def is_prime(value):
    for i in range(2, value):
        prime_value = True
        if value % i == 0:
            prime_value = False
            break 

    if prime_value == True:
        print(value, "This is  a prime value")
    else:
        print(value,"This is not a prime value")
    
is_prime(15) 
is_prime(43)
is_prime(77) 
        
```

__Q12. Dictionary (5 Marks)__

    Input: students = {"Raman": 21, "Pushpkar": 22, "Summer": 0}
    Write a function that returns the name of the oldest student

```python
students = {"Raman":21,"Pushpkar":22,"Karan":23}
age = []

for i in students:
    age.append(students[i])
age.sort(reverse= True)

for i in students:
    if students[i] == age[0]:
        print("Oldest Student in the class is", i)
        break
```

    Input: "Bignalytics"

```python
your_input = input("Enter your word:- ")
word = your_input.lower()
vowels = "aeiou"
count = 0

for i in word:
    if i in vowels:
        count += 1
print("Total volwels in your word ", your_input,"is :- ", count)
```

__Q14. Real-World Mini Project (10 Marks)__

Write a Python program that does the following:

    Input: User enters sales for each day (use while loop, sentinel = -1)
    Calculate:
        Total sales
        Number of days
        Average sales

    Input: Ask customer age for each day, and categorize into:
        <20: “Under 20”
        20–30: “20–30”
        31–50: “31–50”
        >50: “Above 50”

    Output: Print summary table with day-wise sales and age category

🧠 Covers: Input, while loop, if-elif-else, variables, type casting, flow control

```python
a = int(input("Enter the sales:- "))
age = int(input("Enter your age:- "))
days = 1
total_sales = 0
table = [
    ["days","Total_slaes","Age_group"]
]

while a > 0:
    total_sales = a + total_sales
    if age <= 20:
        age_group = "Under_20"
    elif age > 20 and age < 30:
        age_group = "20 - 30"
    elif age > 31 and age < 50:
        age_group = "31 - 50"
    elif age > 50:
        age_group = "Above 50"
    else:
        age_group = "Unknown_age_group"
    
    table.append([days,a,age_group])
    days += 1

    a = int(input("Enter the sales:- "))
    age = int(input("Enter your age:- "))


for row in table:
    print(row)

print("Total_sales is :- ",total_sales)
avg_sales = total_sales / (days - 1 )
print("Avg sales is ",avg_sales)
print("Total_number of days is ", days-1)


```

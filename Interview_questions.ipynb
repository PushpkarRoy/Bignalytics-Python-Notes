Q5. Dictionary – Real Interview Favorite

```python
# Problem
# Given:

employees = {
    "Amit": 45000,
    "Riya": 52000,
    "Pushpkar": 60000
}
```

    Find the employee with highest salary
    Increase salary by 10% if salary > 50,000

📌 Tests: dictionary, loop, condition

```python
employees = {
    "Amit": 45000,
    "Riya": 52000,
    "Pushpkar": 60000
}

salary = []
new_salary = []

for i in employees:
    val = employees[i]
    salary.append(val)

salary.sort(reverse=True)

for key,value in employees.items():
    if value == salary[0]:
        print("Highest salaryed employed is :- ",key)

salary.sort()

for j in salary:
    if j >= 50000:
        calcu = (j * 10) // 100
        final_salary = j + calcu
        new_salary.append(final_salary)
    else:
        new_salary.append(j)


value = 0
for i in employees:
    employees[i] = new_salary[value]
    value += 1

print(employees)



```

6️⃣ Interview-Level String Logic

Function:

    longest_word(sentence)

Rules:

    Ignore punctuation
    Return longest word
    If tie → return first occurrence

⚡ Tests string manipulation skill.

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

longest = ""

for word in words:
    # remove basic punctuation
    word = word.replace(".", "").replace("!", "").replace(",", "").replace("?", "")
    
    if len(word) > len(longest):
        longest = word

print("Longest word is:", longest)
```

8️⃣ Transaction Analyzer (Data Analyst Vibes)

Function:

    transaction_summary(transactions)

Input:

[100, -50, 200, -30, -20]

Return:

    Total income
    Total expense
    Net balance
    Highest expense

⚡ Very real fintech style question.

```python
amount = [100, -50, 200, -30, -20]
income = []
expences = []

for i in amount:
    if i >= 0:
        income.append(i)
    elif i < 0:
        expences.append(i)

total_income = 0 
total_expences = 0 

for j in income:
    total_income = total_income + j 
print("Total income is :- ", total_income)

for k in expences:
    total_expences = total_expences + k 
    final_expence = -total_expences
print("Total expense is :- ", final_expence)

net_balanace = total_income + total_expences
print("Net balance is :- ", net_balanace)

amount.sort()
high_exp = -amount[0]
print("Highest expence is ", high_exp)
```

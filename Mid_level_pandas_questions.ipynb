<h2 style="color:darkorange; font-weight:bold;">Mid-Level Pandas Quesions</h2>

```python
import pandas as pd
```

## Create Data frame 

```python
data = {
    "order_id": [101,102,103,104,105,106,107,108,109,110],
    "customer_name": ["Aman","Riya","Rahul","Sneha","Aman","Karan","Riya","Rahul","Sneha","Aman"],
    "city": ["Delhi","Mumbai","Delhi","Bangalore","Mumbai","Delhi","Bangalore","Mumbai","Delhi","Bangalore"],
    "product": ["Laptop","Phone","Tablet","Laptop","Phone","Tablet","Laptop","Phone","Tablet","Laptop"],
    "category": ["Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics","Electronics"],
    "quantity": [1,2,1,1,3,2,1,2,1,1],
    "price": [60000,20000,30000,65000,21000,28000,62000,19000,31000,64000],
    "order_date": [
        "2024-01-02","2024-01-05","2024-01-07","2024-01-10","2024-01-12",
        "2024-01-15","2024-01-18","2024-01-20","2024-01-23","2024-01-25"
    ]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

<h2 style="color:Red; font-weight:bold;">Practice Questions</h2>

__Q1. Create a new column called revenue.__

```python
df["revenue"] = df["quantity"] * df["price"]
df.head()
```

__Q2. Find the total revenue generated by each city.__

```python
df.groupby("city")["revenue"].sum().sort_values(ascending= True)
```

__Q3. Find the top 3 customers who spent the most money.__

```python
df.groupby("customer_name")["revenue"].sum().sort_values(ascending= False).head(3)
```

__Q4. Find the average order value for each product.__

```python
aov = df.groupby("product")[["revenue","quantity"]].sum()
aov["AOV"] = aov["revenue"] / aov["quantity"]
aov
```

__Q5. Find the total quantity sold for each product.__

```python
df.groupby("product")["quantity"].sum().sort_values(ascending= False)
```

__Q6. Find which city ordered the highest number of products.__

```python
df.head()
```

```python
df.groupby("city")["quantity"].sum().sort_values(ascending= False)
```

__Q7. Find the day with the highest revenue.__

```python
df.groupby("order_date")["revenue"].sum().sort_values(ascending= False).head(5)
```

__Q8. Find customers who placed more than 1 order.__

```python
cust = df.groupby("customer_name")["customer_name"].count().sort_values(ascending= False)
repeatable_customer = cust[cust > 1]
repeatable_customer
```

__Q9. Create a pivot table showing:__

    Rows → city  
    Columns → product  
    Values → total revenue

```python
product_data = pd.pivot_table(
    df,
    index= "city",
    columns= "product",
    values= "revenue",
    aggfunc= "sum"
)
```

```python
product_data
```

__Fill null value__

```python
product_data["Laptop"] = product_data["Laptop"].fillna(0)
```

```python
product_data["Phone"] = product_data["Phone"].fillna(0)
```

```python
product_data["Tablet"] = product_data["Tablet"].fillna(0)
```

```python
product_data
```

__Q10 Find the percentage contribution of each product to total revenue.__

Example output idea:

    Product	          Revenue %
    Laptop	         55%
    Phone	         30%
    Tablet	         15%

```python
df["revenue_percentage"] = (df["revenue"] / df["revenue"].sum()) * 100
```

```python
df["Revenue_Percentage"] = round(((df["revenue"] / df["revenue"].sum()) * 100),2)
df
```

```python
df.groupby("product")["Revenue_Percentage"].sum().astype(str) + "%"
```

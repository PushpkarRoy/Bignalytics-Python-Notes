__1. ATM Withdrawal System__

Write a program:

Input balance and withdrawal amount

If withdrawal > balance → “Insufficient balance”

If withdrawal ≤ balance → “Transaction successful”

```python
name = {
    "Raman":5500,
    "Raj":2200,
    "Monty":7800,
    "Atrik":780,
    "Yash":5800
}
```

```python
user_name = input("Enter your name :- ")

if user_name in name:
    print("Thankyou for login.",user_name)
    user_amount = name[user_name]
    amount = int(input("Enter your withdrawal amount:- "))

    if user_amount >= amount:
        print("Transaction successful")
        print("Thank-you for the withdrawal your withdrawal amount is ", amount)
        print("Your current balance is:- ", user_amount - amount)
        final_amount = user_amount - amount
        name[user_name] = final_amount

    else:
        print("Insufficient balance")

else:
    print("Sorry",user_name,"But your id is not found please write your currect name:- ")
```

__2. Online Shopping Discount__

Rules:

If bill ≥ 5000 → 20% discount

If bill ≥ 3000 → 10% discount

Else → No discount

👉 Print final payable amount.

```python
shopping_bill = int(input("Enter your billing amount:- "))

if shopping_bill >= 3000 and shopping_bill < 5000:
    Discount_amount = (shopping_bill * 10 ) / 100
    print("Your shopping amount is :- ", shopping_bill,)
    print("and your discount amount is:- ", Discount_amount)
    print("Discount 10%")
    print("Your final bill is", shopping_bill - Discount_amount)
elif shopping_bill >= 5000:
    Discount_amount = (shopping_bill * 20 ) / 100
    print("Your shopping amount is :- ", shopping_bill,)
    print("and your discount amount is:- ", Discount_amount)
    print("Discount 20%")
    print("Your final bill is", shopping_bill - Discount_amount)
else:
    print("Final billing amount is", shopping_bill)
    print("sorry but you don't claim any discount")
```

__Q22. Write Python code to count frequency of each element in a list.__

Example:    
    [1,2,2,3,3,3]

```python
list = [1,2,2,2,3,3,3,4]
new_list = []
dec = {}
counting = 1

for i in list:
    if i not in new_list:
        dec[i] = 1
        new_list.append(i)
    elif i in new_list:
        counting = counting + 1
        dec[i] = counting

print(dec)

```

```python
# Other way give some help from Google 
my_list = [1, 2, 2, 2, 3, 3, 3, 4]
dec = {}

for i in my_list:
    if i in dec:
        dec[i] += 1  # If it exists, increment the existing count
    else:
        dec[i] = 1   # If it's new, set it to 1

print(dec)
```

__Q26. Python Data Processing__

You are given a list of employee salaries:

Write a Python program to:

    1. Find the highest salary
    2. Find the lowest salary
    3. Calculate the average salary

Create a new list containing salaries greater than 55000

    Print all results clearly.

```python
salaries = [45000, 52000, 48000, 60000, 75000, 50000, 62000]

salaries.sort()
max = len(salaries) -1

print("The lowest salary is:- ",salaries[0])
print("The highest salary is:- ", salaries[max])
 
for i in salaries:
    i += i 
    avg = i / len(salaries)

avg = round(avg,2)
print("The average value of salary is:-", avg)
```

__Q30. Real Business Problem__

An online store has the following data:

    Customer
    Product
    Price
    Quantity
    City
    Rating

Tasks:

    1. Create Total_Sales = Price × Quantity
    2. Find top 3 customers by spending
    2. Find average rating per product
    4. Find city with highest revenue

```python
customers = ['Aman', 'Sita', 'John', 'Riya', 'Zeeshan', 'Rahul', 'Priya', 'Ananya', 'Vikram', 'Sneha']
products = ['Smartphone', 'Earbuds', 'Charger', 'Smartwatch', 'Powerbank', 'Tablet', 'Case']
cities = ['Indore', 'Mumbai', 'Delhi', 'Pune', 'Bangalore']

# Generating 25 rows of data
data = {
    'Customer': [np.random.choice(customers) for _ in range(25)],
    'Product': [np.random.choice(products) for _ in range(25)],
    'Price': [np.random.choice([15000, 2500, 1000, 5000, 2000, 30000, 500]) for _ in range(25)],
    'Quantity': [np.random.randint(1, 5) for _ in range(25)],
    'City': [np.random.choice(cities) for _ in range(25)],
    'Rating': [np.random.randint(1, 6) for _ in range(25)]
}

store = pd.DataFrame(data)
store.head()
```

```python
# Create total sales 
store["Total_Sales"] = store["Price"] * store["Quantity"]
store.head()
```

```python
# Find top 3 customer by spending 
store.groupby("Customer")["Total_Sales"].sum().sort_values(ascending= False).head(3)
```

```python
# Find average rating per product 
store.groupby("Product")["Rating"].mean().sort_values(ascending= False).round(2)
```

```python
# Find city with highest revenue 
store.groupby("City")["Total_Sales"].sum().sort_values(ascending= False)
```

__Q31. Data Cleaning + Analysis__

Dataset:

    Order_ID
    Customer
    Product
    Price
    Quantity
    Order_Date
    City

But the dataset contains problems:

    Missing values
    Duplicate rows
    Wrong data types

Tasks:

    Remove duplicates
    Handle missing values
    Convert Order_Date to datetime
    Create column Revenue
    Find monthly revenue trend

```python
data = {
    'Order_ID': [101, 102, 102, 103, 104, 105, 105, 106, 107, 108],
    'Customer': ['Aman', 'Sita', 'Sita', 'John', np.nan, 'Riya', 'Riya', 'Rahul', 'Priya', 'Vikram'],
    'Product': ['Laptop', 'Mouse', 'Mouse', 'Monitor', 'Keyboard', 'Laptop', 'Laptop', 'Monitor', np.nan, 'Mouse'],
    'Price': ['55000', '800', '800', '12000', '1500', '52000', '52000', '12500', '900', '800'], # String type (Wrong)
    'Quantity': [1, 2, 2, np.nan, 1, 1, 1, 2, 3, np.nan],
    'Order_Date': ['2024-01-01', '2024-01-02', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08'],
    'City': ['Indore', 'Mumbai', 'Mumbai', 'Delhi', 'Pune', 'Indore', 'Indore', 'Mumbai', 'Delhi', 'Indore']
}

sales = pd.DataFrame(data)
sales 
```

```python
sales.duplicated()
```

```python
# Remove duplicates
sales = sales.drop_duplicates()
```

```python
# Handling missing values 
sales.isnull().sum()
```

```python
# Handle missing value
# 1st Customer name into user
sales["Customer"] = sales["Customer"].fillna("User")
```

```python
# 2nd change product into unknown_product
sales["Product"] = sales["Product"].fillna("Unknown_Product")
```

```python
# 3rd change missing quantity value with the help of mean 
sales["Quantity"] = sales["Quantity"].fillna(sales["Quantity"].mean())
```

```python
# Round of Quantity value
sales["Quantity"] = round(sales["Quantity"])
```

```python
sales.head()
```

```python
# Data cleaning 
# 1. Data type 

sales.dtypes
```

```python
# We saw 2 problem in data type 
# 1. one in price column is object type and we will try to change into int 
# 2. Second one is order date is also object type and we will try to change into datetime 

sales["Price"] = sales["Price"].astype(int)
sales["Order_Date"] = pd.to_datetime(sales["Order_Date"])
sales 
```

```python
# Create column Revenue 
sales["Revenue"] = sales["Price"] * sales["Quantity"]
sales.head()
```

```python
# Create Month_Name column
sales["Month_Name"] = sales["Order_Date"].dt.month_name()
sales.head()
```

```python
# Find Monthly Revenue 
sales.groupby("Month_Name")["Revenue"].sum()
```

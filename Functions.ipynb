Q5. Factorial Function

   a function factorial(n) that returns the factorial of a number using a loop.
   
   Example: factorial(5) → 120

```python
def factorial(n):
    i =  1
    total_value = 1
    for i in range(1,n+ 1,1):
        total_value = i * total_value
    return total_value 

factorial(5)
```

__Q6. Function to Find Maximum__

Write a function find_max(lst) that:

        Takes a list of numbers
        Returns the maximum number

Example: find_max([12, 45, 7, 89, 34]) → 89

```python
def max_finder(*x):
    a = []
    for i in x:
        a.append(i)
    a.sort(reverse= True)
    return ("Max number in this list is ", a[0])


max_finder(12,45,7,89,34)
```

__Q8. Function to Count Vowels__

Write a function count_vowels(text) that:

        Takes a string
        Returns the number of vowels (a, e, i, o, u) in it

Example: "Bignalytics" → 4

```python
def count_vowels(x):
    vowels = "aeiou"
    count = 0
    for i in x:
        if i in vowels:
            count +=1 
    return count

count_vowels("bignalytics")
```
